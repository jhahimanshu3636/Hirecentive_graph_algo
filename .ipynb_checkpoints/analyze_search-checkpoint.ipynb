```python
class Neo4jQueryBuilder:
    @staticmethod
    def build_occupations_nodes_query(row):
        return f"""
        MERGE (c:Occupation {{key_id: '{row.get('ID', '')}'}})  
        ON CREATE SET 
          c.occupation_name = "{row.get('PREFERREDLABEL', '').replace('"','')}",
          c.origin_url = coalesce('{row.get('ORIGINURI', '')}', ""),
          c.code = coalesce('{row.get('CODE', '')}', ""),
          c.description = coalesce("{row.get('DESCRIPTION', '').replace('"','')}", "")
        """

    @staticmethod
    def build_isko_group_node_query(row):
        return f"""
        MERGE (c:ISCO_group {{key_id: '{row.get('ID', '')}'}})  
        ON CREATE SET 
          c.occupation_name = "{row.get('PREFERREDLABEL', '').replace('"','')}",
          c.origin_url = coalesce('{row.get('ORIGINURI', '')}', ""),
          c.code = coalesce('{row.get('CODE', '')}', ""),
          c.description = coalesce("{row.get('DESCRIPTION', '').replace('"','')}", "")
        """

    @staticmethod
    def build_occupation_hierarchy_query(row):
        return f"""
        MATCH (a:Occupation {{key_id: '{row.get('PARENTID', '')}'}}) ,(b:Occupation {{key_id: '{row.get('CHILDID', '')}'}})
        MERGE (a)<-[r:HAS_PARENT_OCCUPATION]-(b)
        """

    @staticmethod
    def build_skill_nodes_query(row):
        return f"""
        MERGE (c:Skill {{key_id: '{row.get('ID', '')}'}})  
        ON CREATE SET 
          c.skill_name = "{row.get('PREFERREDLABEL', '').replace('"','')}",
          c.origin_url = coalesce('{row.get('ORIGINURI', '')}', ""),
          c.skill_type = coalesce('{row.get('SKILLTYPE', '')}', ""),
          c.description = coalesce("{row.get('DESCRIPTION', '').replace('"','')}", "")
        """

    @staticmethod
    def build_skill_occupation_hierarchy(row):
        return f"""
        MATCH (a:Occupation {{key_id: '{row.get('OCCUPATIONID', '')}'}}) ,(b:Skill {{key_id: '{row.get('SKILLID', '')}'}})
        MERGE (a)-[r:HAS_SKILL]->(b)
        SET r.type = '{row.get('RELATIONTYPE', '')}'
        """
```
